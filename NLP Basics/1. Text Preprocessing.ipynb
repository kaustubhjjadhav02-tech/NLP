# Tokenization :
The process of dividing text into individual tokens such as words and phrases is called Tokenisation…

```python
# Natural language toolkit
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
```

```python
tokens = word_tokenize(document)
print(tokens)
```

```python
# 1. See the o/p, every word is separated by whitespace.....
# 2. Even each and every word also separated by punctuation mark comes along with
# that……so this tokenisation is much much better than the split() function…
```

```python
len(tokens)
```

# Stopwords:
Commonly used words (e.g. ‘the’, ‘is’, ‘and’ etc) that are often removed from the text during preprocessing because they carry less meaningful information……

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
print(stop_words)
```

```python
# The above are the set of stopwords, in order to remove those stopwords from the set of tokens we have in our document as
```

```python
tokens = word_tokenize(document)
print(tokens)
```

```python
# Remove bad tokens from the documents
clean_tokens = [token for token in tokens if token not in stop_words]
print(clean_tokens)
```

```python
len(clean_tokens)
```

# Punctuation :
With stopwords, we also want to remove Punctuation…..for this u don’t want ‘nltk library’….
It is directly available in string…..

```python
from string import punctuation
punctuation
```

```python
clean_tokens = [token for token in clean_tokens if token not in punctuation]
print(clean_tokens)
```

```python
len(clean_tokens)
```

# Stemming and Lemmatization

# 1. Stemming:
The process of reducing words to their base form by stripping suffixes, often without considering the context….

```python
stemmer = nltk.PorterStemmer()
```

```python
stemmer.stem("wolves")
```

```python
stemmer.stem("studying")
```

# 2. Lemmatization:
The process of reducing words to their base form or root form by considering the context of the word….

```python
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
```

```python
lemma.lemmatize("wolves")
```

```python
lemma.lemmatize("studying")
```

# Points to remember:
1. Tokenization is mandatory…..
2. Stopwords is a good choice….
3. But Stemming and Lemmatization is optional…...(Stemming u surely ignore but Lemmatization is ur choice….)

```python
# on clean tokens
lemmatized_tokens = [lemma.lemmatize(token) for token in clean_tokens]
print(lemmatized_tokens)
```

```python
clean_tokens = [token for token in clean_tokens if token not in list(punctuation)]
print(clean_tokens)
```

# spacy
import spacy

spaCy is an open-source Python library for **advanced Natural Language Processing (NLP)**, designed specifically for **speed and efficiency in production environments**. It provides pre-trained statistical models and an opinionated API that prioritizes getting real-world NLP applications built quickly and reliably.

---

### Why is it used?

spaCy is used because it provides a fast, efficient, and user-friendly solution for common NLP tasks. Its core design philosophy focuses on performance and production-readiness, unlike other libraries like NLTK which are more suited for academic research and teaching. Key reasons for its use include:

*   **Speed and Efficiency:** It is engineered with Cython to deliver C-like performance when processing large volumes of text data.
*   **Production-Ready:** It is designed to be robust and easy to integrate into real-world, commercial applications.
*   **Integrated & Opinionated:** It offers a single, streamlined pipeline and avoids asking users to choose between multiple algorithms for the same task, simplifying development.
*   **Rich Functionality:** It provides a comprehensive set of capabilities "out-of-the-box," including tokenization, POS tagging, named entity recognition (NER), and dependency parsing.
*   **Deep Learning Integration:** It seamlessly integrates with deep learning frameworks like TensorFlow and PyTorch for advanced custom model training and use of large language models (LLMs).

---

### Where is it used?

spaCy is widely adopted across various industries for real-world applications.

*   **Technology:** Used in the development of chatbots and virtual assistants to parse user queries and understand intent.
*   **Finance & Banking:** Employed for fraud detection, risk assessment, and monitoring compliance by analyzing large volumes of financial documents.
*   **Healthcare & Biotechnology:** Used to analyze medical records and extract valuable insights from research papers to assist in drug discovery and informed decision-making.
*   **E-commerce & Customer Service:** Powers sentiment analysis on customer reviews and feedback, helping businesses improve products and services.
*   **Recruitment:** Used in resume scanning applications to extract relevant skills, experience, and education.
*   **Media & Journalism:** Applied to tasks like topic modeling, entity recognition in articles, and text summarization to process information efficiently.

---

### When is it used?

spaCy should be used when you need to:

*   **Build production-grade NLP applications:** Its focus on performance makes it ideal for real-time applications and large-scale projects.
*   **Process large volumes of text data quickly:** The efficiency built into the library is beneficial for handling big data without slowing down.
*   **Extract structured information from unstructured text:** Its NER and dependency parsing capabilities are perfect for turning raw text into actionable data.
*   **Implement standard NLP tasks with high accuracy:** spaCy's models are highly accurate for common tasks like POS tagging and NER, reducing the need for extensive custom training.
*   **Integrate NLP into existing machine learning pipelines:** Its object-oriented API and custom component support make it easy to integrate with other libraries and systems.
