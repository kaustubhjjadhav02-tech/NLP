```python
text=" I got a great deal on this Laptop, which I have been using for almost a year now. I got it only on Discount but also was able to exchange my old laptop with this! Laptop works really well, it was a good purchase!!"
```

```python
text
```

```python
# Tokenisation
from nltk.tokenize import word_tokenize
words1 = word_tokenize(text)
words1
```

```python
len(words1)
```

```python
# Normalization --> Convert all words to lowercase
words2 = [x.lower() for x in words1]
print(words2)
```

```python
# Stopwords --> remove non-meaningful words
from nltk.corpus import stopwords
stopwords_list = stopwords.words('english')
print(stopwords_list)
len(stopwords_list)
```

```python
words3 = []
for i in words2:
  if i not in stopwords_list:
    words3.append(i)
```

```python
len(words3)
```

```python
print(words3)
```

```python
# Remove Punctuations
import string
string.punctuation
```

```python
words4 = []
for i in words3:
  if i not in string.punctuation:
    words4.append(i)
```

```python
len(words4)
```

```python
print(words4)
```

```python
# Stemmatization and Lemmatization
e_words = ["wait","waiting","waited","waits"]

from nltk.stem import PorterStemmer
p_stemmer = PorterStemmer()

for i in e_words:
  print(i+' -->'+p_stemmer.stem(i))

print()
print()

from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()

for i in e_words:
  print(i+' -->'+Lemmatizer.lemmatize(i))
```

```python
# Stemmatization and Lemmatization
text = ["studies","studying","cries","cry"]
from nltk.stem import PorterStemmer
p_stemmer_1 = PorterStemmer()

for i in text:
  print(i+' -->'+p_stemmer_1.stem(i))

print()
print()

from nltk.stem import WordNetLemmatizer
Lemmatizer_1 = WordNetLemmatizer()

for i in text:
  print(i+' -->'+Lemmatizer_1.lemmatize(i))
```
