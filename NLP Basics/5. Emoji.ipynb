## Remove emoji:
Rather than converting emoji into text, we can remove emojis

```python
def remove_emoji(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')
```

Function remove_emoji removes emojis and other non-ASCII characters from a string by encoding it to ASCII and ignoring errors.

#Issues with the above Approach:
1. It removes all non-ASCII characters, not just emojis. This means accented letters (e.g., √©, √±, √º) and special symbols (e.g., ‚Ç¨, ¬©, ‚Ñ¢) will also be removed.
2. Some emojis are represented as ASCII sequences (e.g., :D, :-)), which this method won't handle.

#Alternative Approach:
If you want to only remove emojis and keep other non-ASCII characters, you can use the emoji library:

```python
import emoji

def remove_emoji(inputString):
    return emoji.replace_emoji(inputString, replace='')
```

```python
text = "Hello üòä, how are you? Caf√©"
print(remove_emoji(text))  # Output: "Hello , how are you? Caf√©"
```
