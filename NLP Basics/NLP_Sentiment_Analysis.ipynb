```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')

# Instantiate the sentiment analyzer
analyzer = SentimentIntensityAnalyzer()

# Create a new column for sentiment
df_final['Sentiment_0'] = df_final['text'].apply(lambda x: analyzer.polarity_scores(x)['compound'])
df_final.head()
```

```python
# Classify each statement as positive or negative based on the sentiment score
df_final['Sentiment'] = df_final['Sentiment_0'].apply(lambda x: 'positive' if x > 0 else 'negative')

# Display the updated DataFrame
print(df_final)

df_final.head()
```

```python
# Normalisation
df_final["text"] = df_final["text"].str.lower()
df_final.head()
```

```python
# Remove Punctuations
import string
df_final['text'] = df_final['text'].str.replace('[{}]'.format(string.punctuation), '')
df_final.head()
```

```python
# Apply Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()


df_final['text'] = df_final['text'].apply(lambda x: " ".join([lemmatizer.lemmatize(word) for word in x.split()]))
df_final.head()
```

```python
# Remove STOPWORDS
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

df_final['text'] = df_final['text'].apply(lambda x: " ".join([word for word in x.split() if word not in stop_words]))
df_final.head()
```

```python
# Data Partition
Y = df_final["Sentiment"]
x = df_final['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
#vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(x)
x_new = X.toarray()

x_new.shape
```

```python
x_new
```

```python
# Model Building (Logistic Regression)
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(x_new, Y)

# Model Prediction
y_pred = model.predict(x_new)

# Evaluation metrics
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y, y_pred)
import numpy as np
print("accuacy score: ",np.round(accuracy,2))
```

```python
# Model Building (Naive Bayes)
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()
model.fit(x_new, Y)

# Model Prediction
y_pred = model.predict(x_new)

# Evaluation metrics
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(Y, y_pred)
import numpy as np
print("accuacy score: ",np.round(accuracy,2))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
# Create a CountVectorizer object with ngram_range=(2, 2) to extract bigrams
vect = CountVectorizer(ngram_range=(2, 2))

# Fit the vectorizer to the document
counts = vect.fit_transform(x)

# Get the vocabulary of the vectorizer
vocab = vect.get_feature_names_out()

# Get the top 20 bigram counts
top_20_bigrams = counts.toarray().sum(axis=0).argsort()[-20:]

# Create a bar chart of the top 20 bigram counts
plt.figure(figsize=(15, 7))
plt.bar(vocab[top_20_bigrams], counts.toarray()[0, top_20_bigrams])
plt.xticks(rotation=90)
plt.xlabel("Bigrams")
plt.ylabel("Count")
plt.title("Top 20 Bigram Counts")
plt.show()
```
