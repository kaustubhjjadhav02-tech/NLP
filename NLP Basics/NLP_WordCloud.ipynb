# Importing dataset

```python
# specify the path to your .txt file
file_path = "apple.txt"

# open and read the contents of the file
with open(file_path, 'r', encoding= 'utf-8') as file:
  text_data = file.read()
```

## DATA PREPROCESSING

```python
import nltk
nltk.download('all')
```

```python
# Tokenize the text
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
words = word_tokenize(text_data)

print(words)
len(words)
```

```python
words
```

```python
# Normalization --> Convert all words to lowercase
words2 = [x.lower() for x in words]

# Print the words
print(words2)
```

```python
# STOPWORD --> remove non-meaningful words
from nltk.corpus import stopwords
stopwords_list = stopwords.words('english')
stopwords_list
len(stopwords_list)

# Remove stop words
words3 = [x for x in words2 if x not in stopwords_list]
words3
print("after stopwords:", len(words3))
```

```python
2972/4793
```

```python
# Remove Punctuation
import string
string.punctuation
punctuations = string.punctuation
punctuations
```

```python
words_without_punctuation = []

for x in words3:
    if x not in punctuations:
        words_without_punctuation.append(x)

print(words_without_punctuation)
print('words without punctuation:',len(words_without_punctuation))
words_without_punctuation[:5]
```

```python
# Stemmatization
from nltk.stem import PorterStemmer
porter_stemmer = PorterStemmer()

stemmed_words = [porter_stemmer.stem(i) for i in words_without_punctuation]
print(len(stemmed_words))
```

```python
# Lemmatization
from nltk.stem import WordNetLemmatizer
Lemmatizer = WordNetLemmatizer()

lemma_words = [Lemmatizer.lemmatize(x) for x in words_without_punctuation]

print(len(lemma_words))
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

#words4 = [x for x in lemma_words if x not in ["apple", "product", "air", "laptop",'macbook','mac','work','use','pro']]

# Create a dictionary to store the word frequency
word_counts = {}
for x in stemmed_words:
  if x not in word_counts:
    word_counts[x] = 0
  word_counts[x] += 1

word_counts
```

```python
# Create the word cloud
wordcloud = WordCloud(width=800, height=600, background_color="white").generate_from_frequencies(word_counts)

# Display the word cloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
doc = " ".join(words4)
doc
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer object with ngram_range=(2, 2) to extract bigrams
vect = CountVectorizer(ngram_range=(2, 2))

# Fit the vectorizer to the document
counts = vect.fit_transform([doc])

# Get the vocabulary of the vectorizer
vocab = vect.get_feature_names_out()

# Get the top 20 bigram counts
top_20_bigrams = counts.toarray().sum(axis=0).argsort()[-20:]

# Create a bar chart of the top 20 bigram counts
plt.figure(figsize=(15, 7))
plt.bar(vocab[top_20_bigrams], counts.toarray()[0, top_20_bigrams])
plt.xticks(rotation=90)
plt.xlabel("Bigrams")
plt.ylabel("Count")
plt.title("Top 20 Bigram Counts")
plt.show()
```
